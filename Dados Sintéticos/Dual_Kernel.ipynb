```python
import numpy as np
import matplotlib.pyplot as plt

def plot_2d_data(X, y, title="Dataset"):
    # Create a scatter plot
    plt.figure(figsize=(8, 6))
    labels = np.unique(y)
    for label in labels:
        mask = y == label
        plt.scatter(X[mask, 0], X[mask, 1], label=f"Label {label}", alpha=0.7)

    plt.xlabel("X1")
    plt.ylabel("X2")
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def plot_decision_boundary(ecoc, BD, y):
    h = 0.02
    x_min, x_max = BD[:, 0].min() - 1, BD[:, 0].max() + 1
    y_min, y_max = BD[:, 1].min() - 1, BD[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    grid = np.c_[xx.ravel(), yy.ravel()]
    Z = ecoc.predict(grid)
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
    plt.scatter(BD[:, 0],BD[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors="k")
    plt.title("...")
    plt.xlabel("X1")
    plt.ylabel("X2")
    plt.show()
    
```

$$\text{Implementao}$$

```python
class KernelLBClassifierGD:
    def __init__(self, degree=2, gamma=1.0, coef=1.0, lambda_=0.1, lr=0.01, max_iter=1000):
        self.degree = degree
        self.gamma = gamma
        self.coef = coef
        self.lambda_ = lambda_
        self.lr = lr
        self.max_iter = max_iter

    def kernel(self, X1, X2):
        return (self.gamma * X1 @ X2.T + self.coef) ** self.degree

    def rbf_kernel(self,X1, X2):
        # ||x - x'||^2 = ||x||^2 + ||x'||^2 - 2x, x'
        X1_sq = np.sum(X1**2, axis=1).reshape(-1, 1)  
        X2_sq = np.sum(X2**2, axis=1).reshape(1, -1)  
        sq_dists = X1_sq + X2_sq - 2 * X1 @ X2.T
        return np.exp(-self.gamma * sq_dists)

    def linear_kernel(self,X1,X2):
        return X1 @ X2.T

    def fit(self, X, y):
        y = np.where(y == 1, 1, -1)
        n = X.shape[0]
        self.X_train = X
        K = self.kernel(X, X)
        self.alpha = np.zeros(n)

        for _ in range(self.max_iter):
            f = K @ self.alpha
            p = 1 / (1 + np.exp(y * f))  
            grad = -K.T @ (y * p) + self.lambda_ * self.alpha
            self.alpha -= self.lr * grad

    def predict_proba(self, X):
        K_test = self.kernel(X, self.X_train)
        scores = K_test @ self.alpha
        return 1 / (1 + np.exp(-scores))

    def predict(self, X):
        return (self.predict_proba(X) > 0.5).astype(int)
```

$\text{ECOC}$

```python
class ECOCClassifier:
    def __init__(self, code_matrix, **kernel_params):
        self.code_matrix = code_matrix
        self.kernel_params = kernel_params
        self.classifiers = []

    def fit(self, X, y):
        n_classes, n_classifiers = self.code_matrix.shape
        self.classifiers = []

        for j in range(n_classifiers):
            binary_y = np.array([self.code_matrix[label, j] for label in y])
            clf = KernelLBClassifierGD(**self.kernel_params)
            clf.fit(X, binary_y)
            self.classifiers.append(clf)

    def predict(self, X):
        predictions = np.array([clf.predict(X) for clf in self.classifiers]).T
        pred_classes = []
        for row in predictions:
            distances = np.sum(row != self.code_matrix, axis=1)
            pred_classes.append(np.argmin(distances))
        return np.array(pred_classes)
```

$$\text{Mini Batch}$$

```python
class KernelLBClassifierMB:
    def __init__(self, degree=2, gamma=1.0, coef=1.0,lambda_=0.1, lr=0.01, max_iter=1000, batch_size=None):
        self.degree = degree
        self.gamma = gamma
        self.coef = coef
        self.lambda_ = lambda_
        self.lr = lr
        self.max_iter = max_iter
        self.batch_size = batch_size  

    def kernel(self, X1, X2):
        return (self.gamma * X1 @ X2.T + self.coef) ** self.degree

    def rbf_kernel(self,X1, X2):
        # ||x - x'||^2 = ||x||^2 + ||x'||^2 - 2x, x'
        X1_sq = np.sum(X1**2, axis=1).reshape(-1, 1)  
        X2_sq = np.sum(X2**2, axis=1).reshape(1, -1)  
        sq_dists = X1_sq + X2_sq - 2 * X1 @ X2.T
        return np.exp(-self.gamma * sq_dists)

    def linear_kernel(self,X1,X2):
        return X1 @ X2.T
        
    def fit(self, X, y):
        y = np.where(y == 1, 1, -1)
        self.X_train = X
        self.y_train = y
        n = len(y)
        K = self.kernel(X, X)
        self.alpha = np.zeros(n)

        for _ in range(self.max_iter):
            if self.batch_size is None:
                idx = np.arange(n)
            else:
                idx = np.random.choice(n, self.batch_size, replace=False)

            K_batch = K[idx][:, idx]        
            alpha_batch = self.alpha[idx]
            y_batch = y[idx]

            f = K_batch @ alpha_batch
            p = 1 / (1 + np.exp(y_batch * f))  # sigmoid(-y*f)
            grad = -K[idx][:, idx].T @ (y_batch * p) + self.lambda_ * alpha_batch
            self.alpha[idx] -= self.lr * grad

    def predict_proba(self, X):
        K_test = self.kernel(X, self.X_train)
        scores = K_test @ self.alpha
        return 1 / (1 + np.exp(-scores))

    def predict(self, X):
        return (self.predict_proba(X) > 0.5).astype(int)
```

$$\text{Sequencial}$$

```python
class KernelLBClassifierSGD:
    def __init__(self, degree=2, gamma=1.0, coef=1.0,lambda_=0.1, lr=0.01, max_iter=1000):
        self.degree = degree
        self.gamma = gamma
        self.coef = coef
        self.lambda_ = lambda_
        self.lr = lr
        self.max_iter = max_iter 

    def kernel(self, X1, X2):
        return (self.gamma * X1 @ X2.T + self.coef) ** self.degree

    def rbf_kernel(self,X1, X2):
        # ||x - x'||^2 = ||x||^2 + ||x'||^2 - 2x, x'
        X1_sq = np.sum(X1**2, axis=1).reshape(-1, 1)  
        X2_sq = np.sum(X2**2, axis=1).reshape(1, -1)  
        sq_dists = X1_sq + X2_sq - 2 * X1 @ X2.T
        return np.exp(-self.gamma * sq_dists)

    def linear_kernel(self,X1,X2):
        return X1 @ X2.T

    def fit(self, X, y):
        y = np.where(y == 1, 1, -1)
        self.X_train = X
        self.y_train = y
        n = len(y)
        K = self.kernel(X, X)
        self.alpha = np.zeros(n)

        for epoch in range(self.max_iter):
            for i in range(n):  # sequential, not random
                f_i = K[i] @ self.alpha
                p_i = 1 / (1 + np.exp(y[i] * f_i)) 
                grad_i = -y[i] * p_i * K[i, i] + self.lambda_ * self.alpha[i]
                self.alpha[i] -= self.lr * grad_i

    def predict_proba(self, X):
        K_test = self.kernel(X, self.X_train)
        scores = K_test @ self.alpha
        return 1 / (1 + np.exp(-scores))

    def predict(self, X):
        return (self.predict_proba(X) > 0.5).astype(int)
```
